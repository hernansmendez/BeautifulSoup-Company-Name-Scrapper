```python
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementClickInterceptedException
from selenium import webdriver

# Launch the web browser using Selenium
driver = webdriver.Chrome()  # or webdriver.Firefox() for Firefox

# Navigate to the webpage using Selenium
driver.get("https://www.multiunitfranchisingconference.com/page/2210310/exhibitors")

# Find the container that holds the logos of the partners/exhibitors
partners_container = driver.find_element(By.CLASS_NAME, "partners-container")

# Find all the logos in the container and extract the company name
df = pd.DataFrame(columns=["company_name"])
for logo in partners_container.find_elements(By.CLASS_NAME,"partner-logo"):
    # Click on the logo element to open the pop-up window
    while True:
        try:
            logo.click()
            break
        except ElementClickInterceptedException:
            time.sleep(0.1)

    # Wait for the pop-up window to load
    time.sleep(0.1)  # adjust the sleep time as needed

    # Extract the company name text from the pop-up window using Selenium
    try:
        company_name = driver.find_element(By.XPATH, "/html/body/div[9]/div[3]/div/div[1]/h3").text
    except:
        company_name = "Error: Company name not found"

    driver.find_element(By.TAG_NAME,'svg').click()
    
    df = pd.concat([df, pd.DataFrame({'company_name': [company_name]})], ignore_index=True)
    
    #print(company_name)

driver.close()

print(len(df.company_name))

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df.head()
```
